```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(4)

# Generate random ages with a normal distribution
mean_age = 40
std_dev_age = 15
num_samples = 150  # Change the number of data points

ages = np.random.normal(mean_age, std_dev_age, num_samples)
ages = np.clip(ages, 0, 80)  # Clip ages to be between 0 and 80

# Set a constant y-value for the horizontal line
horizontal_line_y = 0

# Plot the scatter plot with constant y-value
plt.scatter(ages, np.full_like(ages, horizontal_line_y), color='red')

plt.ylim(-0.1, 0.1)

# Set labels and title
plt.xlabel('Age')
plt.title('Ages with Circles')

# Show the plot
plt.show()
```

```python
# Specify bins explicitly
bins = np.linspace(0, 80, 10)  # 9 bins from 0 to 80

# Create histogram with circles
plt.hist(ages, bins=bins, edgecolor='black', alpha=0.7, weights=np.zeros_like(ages))

# Get histogram counts and bin edges
counts, bin_edges = np.histogram(ages, bins=bins)

# Plot circles for each bin
for i in range(len(bin_edges) - 1):
    bin_center = (bin_edges[i] + bin_edges[i + 1]) / 2
    plt.scatter([bin_center] * counts[i], np.arange(counts[i]), marker='o', color='red', alpha=0.7)

# Set labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages with Circles')

# Show the plot
plt.show()
```

```python
# Set the seed for reproducibility
np.random.seed(42)

# Parameters for the normal distribution
mean = 5  # Mean of the distribution
std_dev = 2  # Standard deviation of the distribution

# Number of data points
num_points = 1000

# Generate a dataset with a normal distribution and small noise
data = np.random.normal(mean, std_dev, num_points) + np.random.normal(0, 0.5, num_points)

# Plot the histogram
plt.hist(data, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')

# Plot the theoretical normal distribution for comparison
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = (1/(std_dev * np.sqrt(2 * np.pi))) * np.exp(-(x - mean)**2 / (2 * std_dev**2))
plt.plot(x, p, 'k', linewidth=2)

# Add labels and title
plt.title('Histogram of Data with Normal Distribution and Small Noise')
plt.xlabel('Value')
plt.ylabel('Density')

# Show the plot
plt.show()
```

```python
from scipy.stats import binom

# Set the parameters for the binomial distribution
n = 10  # Number of trials
p = 0.5  # Probability of success in each trial

# Generate a range of x values (number of successes)
x_values = np.arange(0, n+1)

# Calculate the probability mass function (PMF) for each x value
pmf_values = binom.pmf(x_values, n, p)

# Plot the binomial distribution
plt.bar(x_values, pmf_values, align='center', alpha=0.7)
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

```python
from scipy.special import comb

# Set the parameters for the binomial distribution
n = 10  # Number of trials
p = 0.3  # Probability of success in each trial

# Generate a range of x values (number of successes)
x_values = np.arange(0, n+1)

# Calculate the probability mass function (PMF) for each x value
pmf_values = [comb(n, k) * (p ** k) * ((1 - p) ** (n - k)) for k in x_values]

# Plot the binomial distribution
plt.bar(x_values, pmf_values, align='center', alpha=0.7)
plt.title('Binomial Distribution (n=10, p=0.3)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

```python
from scipy.stats import norm

# Parameters for the first normal distribution
mean1 = 0
std_dev1 = 1

# Parameters for the second normal distribution
mean2 = 3
std_dev2 = 1.5

# Generate x values
x = np.linspace(-5, 8, 1000)

# Calculate PDFs for each distribution
pdf1 = norm.pdf(x, mean1, std_dev1)
pdf2 = norm.pdf(x, mean2, std_dev2)

# Plot the PDFs as lines
plt.plot(x, pdf1, label='Distribution 1')
plt.plot(x, pdf2, label='Distribution 2')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Two Normal Distributions - Probability Density Functions')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Define the mean and standard deviation
mean = 0
std_dev = 1

# Define the normal distribution function
def normal_distribution(x):
    return norm.pdf(x, mean, std_dev)

# Define the cumulative distribution function
def normal_cdf(x):
    return norm.cdf(x, mean, std_dev)

# Define the integration bounds
lower_bound = -1
upper_bound = 1

# Calculate the probability using CDF
probability_lower = normal_cdf(lower_bound)
probability_upper = normal_cdf(upper_bound)

# Calculate the area under the curve using CDF
area_under_curve = probability_upper - probability_lower

# Generate x values for plotting
x_values = np.linspace(-3, 3, 1000)
y_values_pdf = normal_distribution(x_values)
y_values_cdf = normal_cdf(x_values)

# Plot the normal distribution curve and CDF
plt.figure(figsize=(10, 6))

# Plot PDF
plt.subplot(2, 1, 1)
plt.plot(x_values, y_values_pdf, label='PDF')
plt.fill_between(x_values, y_values_pdf, where=[(x >= lower_bound and x <= upper_bound) for x in x_values], alpha=0.3, label='Area under PDF')
plt.title('Normal Distribution PDF and Area under the Curve')
plt.xlabel('X-axis')
plt.ylabel('Probability Density Function')
plt.legend()

# Plot CDF
plt.subplot(2, 1, 2)
plt.plot(x_values, y_values_cdf, label='CDF')
plt.axvline(lower_bound, color='red', linestyle='--', label=f'Lower Bound (x={lower_bound})\nProbability: {probability_lower:.4f}')
plt.axvline(upper_bound, color='blue', linestyle='--', label=f'Upper Bound (x={upper_bound})\nProbability: {probability_upper:.4f}')
plt.fill_between(x_values, y_values_cdf, where=[(x >= lower_bound and x <= upper_bound) for x in x_values], alpha=0.3, label='Area under CDF')
plt.title('Normal Distribution CDF and Probability Calculation')
plt.xlabel('X-axis')
plt.ylabel('Cumulative Distribution Function')
plt.legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()

# Display the calculated area using CDF
print(f'Area under the curve between x={lower_bound} and x={upper_bound} using CDF: {area_under_curve:.4f}')
```

```python
from scipy.integrate import quad

# Define the mean and standard deviation
mean = 0
std_dev = 1

# Define the normal distribution function
def normal_distribution(x):
    return norm.pdf(x, mean, std_dev)

# Define the integration bounds
lower_bound = -2
upper_bound = 2

# Generate x values for plotting
x_values = np.linspace(-3, 3, 1000)
y_values_pdf = normal_distribution(x_values)

# Plot the normal distribution curve and CDF
plt.figure(figsize=(10, 6))

# Plot PDF
plt.subplot(2, 1, 1)
plt.plot(x_values, y_values_pdf, label='PDF')
plt.fill_between(x_values, y_values_pdf, where=[(x >= lower_bound and x <= upper_bound) for x in x_values], alpha=0.3, label='Area under PDF')
plt.title('Normal Distribution PDF and Area under the Curve')
plt.xlabel('X-axis')
plt.ylabel('Probability Density Function')
plt.legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()

# Use quad to integrate the PDF over the specified range
area_under_curve, _ = quad(normal_distribution, lower_bound, upper_bound)

print(f"Area under the curve between -2 and 2 (using PDF): {area_under_curve}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for demonstration
np.random.seed(42)
x = np.random.rand(4) * 10
y = 2 * x + 1 + np.random.randn(4) * 3  # Linear relation with some noise

# Perform linear regression to get the line of best fit
coefficients = np.polyfit(x, y, 1)
line_of_best_fit = np.polyval(coefficients, x)

# Calculate SSR points
ssr_points = y - line_of_best_fit

# Plot the data points
plt.scatter(x, y, label='Data Points')

# Plot the line of best fit
plt.plot(x, line_of_best_fit, color='red', label='Linear Fit')

# Annotate SSR points
for i in range(len(x)):
    plt.annotate(f'SSR: {ssr_points[i]:.2f}', (x[i], y[i]), textcoords="offset points", xytext=(0, 10), ha='center')

    # Annotate real value vs predicted value
    plt.annotate(f'Real: {y[i]:.2f}\nPredicted: {line_of_best_fit[i]:.2f}', (x[i], y[i]),
                 textcoords="offset points", xytext=(0, -20), ha='center', color='blue')

    # Add annotation lines from real to predicted with green lines
    plt.plot([x[i], x[i]], [y[i], line_of_best_fit[i]], color='green', linestyle='--')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data points
np.random.seed(42)
X = np.random.rand(10) * 10
y = 2 * X + 1 + np.random.randn(10) * 2  # Linear relationship with some noise

# Define the function to calculate SSR
def calculate_ssr(X, y, slope, intercept):
    y_pred = slope * X + intercept
    residuals = y - y_pred
    return np.sum(residuals ** 2)

# Initialize variables
slope = 1.5  # Initial slope
intercept = -2  # Initial intercept

# Plot initial data
plt.scatter(X, y, label='Data points')

# Plot initial linear line
plt.plot(X, slope * X + intercept, color='red', label='Initial Line')

# Calculate and print initial SSR
initial_ssr = calculate_ssr(X, y, slope, intercept)
print(f'Initial SSR: {initial_ssr:.2f}')

# Update slope and intercept to fit the data better
for i in range(3):
    slope += 0.4
    intercept += 0.5

    # Plot updated linear line
    plt.plot(X, slope * X + intercept, label=f'Line {i+1}')

    # Calculate and print SSR
    current_ssr = calculate_ssr(X, y, slope, intercept)
    print(f'Line {i+1} SSR: {current_ssr:.2f}')

# Show the plot
plt.legend()
plt.xlabel('X')
plt.ylabel('y')
plt.title('SSR with Fitting Linear Lines')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data points
np.random.seed(42)
X = np.random.rand(10) * 10
y = 2 * X + 1 + np.random.randn(10) * 2  # Linear relationship with some noise

# Define the function to calculate SSR
def calculate_ssr(X, y, slope, intercept):
    y_pred = slope * X + intercept
    residuals = y - y_pred
    return np.sum(residuals ** 2)

# Initialize variables
slope = 1.5  # Initial slope
intercept = -2  # Initial intercept

# Plot initial data
fig, axs = plt.subplots(2, 1, figsize=(8, 10))
axs[0].scatter(X, y, label='Data points')

# Plot initial linear line
axs[0].plot(X, slope * X + intercept, color='red', label='Initial Line')

# Calculate and print initial SSR
initial_ssr = calculate_ssr(X, y, slope, intercept)
print(f'Initial SSR: {initial_ssr:.2f}')

# Lists to store SSR values
ssr_values = [initial_ssr]

# Update slope and intercept to fit the data better
for i in range(5):
    slope += 0.4
    intercept -= 0.3

    # Plot updated linear line
    axs[0].plot(X, slope * X + intercept, label=f'Line {i+1}')

    # Calculate and print SSR
    current_ssr = calculate_ssr(X, y, slope, intercept)
    print(f'Line {i+1} SSR: {current_ssr:.2f}')

    # Append SSR value to the list
    ssr_values.append(current_ssr)

# Plot SSR values
axs[1].plot(range(1, len(ssr_values)+1), ssr_values, marker='o', color='green')
axs[1].set_xlabel('Iteration')
axs[1].set_ylabel('SSR')
axs[1].set_title('SSR Decrease with Fitting Linear Lines')

# Show the plot
axs[0].legend()
axs[0].set_xlabel('X')
axs[0].set_ylabel('y')
axs[0].set_title('Fitting Linear Lines')
plt.tight_layout()
plt.show()
```
