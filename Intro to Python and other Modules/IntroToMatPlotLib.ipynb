***Figure***

The whole figure. The Figure keeps track of all the child Axes, a group of 'special' Artists (titles, figure legends, colorbars, etc), and even nested subfigures.

```python
import matplotlib.pyplot as plt
import numpy as np

#fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
ax.plot([1,2,3,4], [1,3,4,2]) #plot some data on the axes
```

```python
np.random.seed(475) #seed the random number generator (any number)

data = {'xData': np.arange(50),
        'color': np.random.randint(0, 50, 50),
        'size': np.random.randn(50) * 100
        }

data['result'] = data['xData'] + 20 * np.random.randn(50)

fig2, ax2 = plt.subplots(figsize=(5, 2.7))  # a figure with a single Axes
ax2.scatter('xData', 'result', c = 'color', s = 'size', data = data)
ax2.set_title('My Figure')
ax2.set_xlabel('xData')
ax2.set_ylabel('Result')
```

```python
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, x, label='linear', color='blue', linewidth=3)  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic', color='orange', linewidth=5, linestyle='--')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic', color='green', linewidth=1)  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.annotate('Here', xy=(2,7),  xytext=(1.75, 6), arrowprops=dict(facecolor='black', shrink=0.05))
ax.legend();  # Add a legend.
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.scatter(x, x**2, s=50, facecolor='C0', edgecolor='k');

mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(5, 2.7))
# the histogram of the data
n, bins, patches = ax.hist(x, 50, density=True, facecolor='C0', alpha=0.75)

ax.set_xlabel('Length [cm]')
ax.set_ylabel('Probability')
ax.set_title('Aardvark lengths\n (not really)')
ax.text(75, .025, r'$\mu=115,\ \sigma=15$')
ax.axis([55, 175, 0, 0.03])
ax.grid(True);
```

```python
n_bins = 20

# Generate two normal distributions
dist1 = np.random.normal(170, 10, 250)
dist2 = 0.4 * dist1 + 5

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

# We can set the number of bins with the *bins* keyword argument.
axs[0].hist(dist1, bins=n_bins)
axs[1].hist(dist2, bins=n_bins)
```
